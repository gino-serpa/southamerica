# Question: For authors that publish in both databases what is their database publication history?

```python
import swl
from pathlib import Path
from pprint import pprint
```

## Choose the datasets (epidemiology)

```python
scielo_files = Path("Data_All")/'epidemiology-scielo+2'
wos_files = Path("Data_All")/'epidemiology-wos'
```

```python
# Read the data
scielo_df = swl.ingest_folder('scielo', data_folder=scielo_files)
wos_df    = swl.ingest_folder('wos',    data_folder=wos_files)
```

```python
len(wos_df), len(scielo_df)
```

## Create the dicts

```python
authors_scielo, papers_scielo, institutions_scielo = swl.get_scielo_dicts(scielo_df)
authors_wos, papers_wos, institutions_wos = swl.get_wos_dicts(wos_df)
```

```python
print(len(authors_scielo), len(papers_scielo),len(institutions_scielo))
print(len(authors_wos), len(papers_wos),len(institutions_wos))
```

# Find the authors intersection

```python
authors_scielo_set = set(list(authors_scielo.keys()))
authors_wos_set = set(list(authors_wos.keys()))
```

```python
authors_intersection = authors_scielo_set.intersection(authors_wos_set)
len(authors_wos_set),len(authors_scielo_set),len(authors_intersection)
```

```python
author_int_list = list(authors_intersection)
```

```python
author_int_list
```

## Now Try to answer the question

```python
authors_history = swl.get_authors_history(author_int_list, 
                                          authors_wos, papers_wos,
                                          authors_scielo, papers_scielo)
```

```python
global_history =swl.get_global_history(authors_history)
```

```python
global_history
```

```python
from bokeh.plotting import figure, show
```

```python
from bokeh.io import output_notebook
```

```python
output_notebook()
history_plot=figure()

scielo_years = sorted([year for year in global_history['scielo']])
print(scielo_years)
scielo_output = [global_history['scielo'][year] for year in scielo_years]
print(scielo_output)
history_plot.line(scielo_years, scielo_output,color='red', legend_label='scielo', line_width=2)

overlap_years = sorted([year for year in global_history['overlap']])
print(overlap_years)
overlap_output = [global_history['overlap'][year] for year in overlap_years]
print(overlap_output)
history_plot.line(overlap_years, overlap_output, color='green',legend_label='overlap',line_width=2)

wos_years = sorted([year for year in global_history['wos']])
print(wos_years)
wos_output = [global_history['wos'][year] for year in wos_years]
print(wos_output)
history_plot.line(wos_years, wos_output, color='blue',legend_label='wos', line_width=2)

show(history_plot)
```
